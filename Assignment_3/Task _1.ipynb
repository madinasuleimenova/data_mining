```python
# Libs that I use for this task
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
import sklearn.neural_network
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
```

```python
dataset = load_wine()
x = dataset.data
y = dataset.target
```

```python
# Now I can devide my data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Now wwe can implement some classification algorithms 

# For example Artificial neural networks(ANN)
# I read about it in the link 
# We can create instance in the next way
neural_network = sklearn.neural_network.MLPClassifier(hidden_layer_sizes=(10,10), activation='relu')
# Now we can fit our model to the training data
neural_network.fit(x_train, y_train)
# And after that we can start predicting our target
y_pred = neural_network.predict(x_test)
```

```python
# To see is it accurate or not
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
# As we can see its accuracy score is very low 
```

```python
# Now lets create another classification algorithm
# For example Naive Bayes

#GaussianNB : This classifier is used for data that is distributed normally.
# MultinomialNB : This classifier is used for data that is counts of occurrences.
# BernoulliNB : This classifier is used for data that is binary (0 or 1).

# I decided to use GaussianNb => because there are continuouse numeric variables
clf2 = GaussianNB()
clf2.fit(x_train, y_train)
y_pred2 = clf2.predict(x_test)
```

```python
accuracy2 = accuracy_score(y_test, y_pred2)
print(accuracy2)
```

```python
# Libs that I used for this exercise
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
import matplotlib.pyplot as plt
```

```python
# Okay, here we need to create confusion matrix and classification report
conf_mat = confusion_matrix(y_test, y_pred)
print(conf_mat)
```

```python
conf_mat_display = ConfusionMatrixDisplay(confusion_matrix = conf_mat, display_labels = [0,1,2])
```

```python
conf_mat_display.plot()
plt.show()
```

```python
# Now we can create classification report
report = classification_report(y_test, y_pred)
```
