###  Exercise 5: Implementing Hierarchical Clustering

```python
# Libs that i use
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
```

```python
# First of all lets download our dataset
dataset = pd.read_csv('C:\\Users\\sulej\\OneDrive\\Рабочий стол\\7 semester\\Data Mining\\Mall_Customers.csv')
```

```python
# Now I can choose features by which I will cluster
x = dataset[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# Plot a dendrogram to visualize the hierarchy
plt.figure(figsize=(10, 6))
dendrogram = sch.dendrogram(sch.linkage(x_scaled, method='ward'))
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Samples')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
agglom = AgglomerativeClustering(n_clusters=5)  # Using 5 clusters for comparison
y_agglom = agglom.fit_predict(x_scaled)
```

```python
# Visualize the clusters
plt.figure(figsize=(10, 6))
plt.scatter(x_scaled[:, 0], x_scaled[:, 1], c=y_agglom, cmap='viridis', s=50)
plt.title('Agglomerative Clustering on Mall Customer Segmentation Dataset')
plt.xlabel('Annual Income (scaled)')
plt.ylabel('Spending Score (scaled)')
plt.show()
```

### Exercise 6: Introduction to DBSCAN for the New Dataset

```python
from sklearn.cluster import DBSCAN

# Adjust eps and min_samples as needed
dbscan = DBSCAN(eps=0.5, min_samples=5)
y_dbscan = dbscan.fit_predict(x_scaled)
```

```python

# Visualize the clusters
plt.figure(figsize=(10, 6))
plt.scatter(x_scaled[:, 0], x_scaled[:, 1], c=y_dbscan, cmap='viridis', s=50)
plt.title('DBSCAN Clustering on Dataset')
plt.xlabel('Ann income (scaled)')
plt.ylabel('Spend score (scaled)')
plt.show()
```
