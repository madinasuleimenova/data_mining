####  Feature Importance with Random Forest

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = load_wine()
```

```python
# Now we can split our data into training and testing sets
# Lets define our features and target
x = dataset.data
y = dataset.target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# Now we can train Random Forest Classifier
RandomForest = RandomForestClassifier(n_estimators=100, random_state=42)
RandomForest.fit(x_train, y_train)
```

```python
# Now we can extract feature importances
importances = RandomForest.feature_importances_
feature_names = dataset.feature_names
```

```python
# Now lets visualize
```

```python
indices = np.argsort(importances)[::-1]  # Sort features by importance
plt.figure(figsize=(10, 6))
plt.title("Feature Importances in the Wine Dataset")
plt.bar(range(x.shape[1]), importances[indices], align="center")
plt.xticks(range(x.shape[1]), [feature_names[i] for i in indices], rotation=90)
plt.xlabel("Feature")
plt.ylabel("Importance")
plt.tight_layout()
plt.show()
```
