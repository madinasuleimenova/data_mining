#### Decision Tree Regression

```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Firstly lets load our dataset
dataset = load_boston()
```

```python
# Now we can split this dataset into training and testing sets
x = dataset.data
y = dataset.target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Now we can start training our model
dec_tree = DecisionTreeRegressor()
dec_tree.fit(x_train, y_train)
```

```python
# Now lets predict our teaget
y_pred = dec_tree.predict(x_test)
```

```python
# Now we can evaluate our model
mse = mean_squared_error(y_test, y_pred)
print(mse)
```

```python
plt.figure(figsize=(12, 8))
plot_tree(dec_tree, filled=True, feature_names=dataset.feature_names)
plt.show()
```
